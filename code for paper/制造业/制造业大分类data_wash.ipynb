```python
def edu_trans(x):
    if x in ['男\u3000 女\u3000 計\n学\u3000 歴   計','     男\n学\u3000 歴   計','     女\n学\u3000 歴   計']:
        return 'delete'
    elif x =='中   学   卒':
        return 9
    elif x == '高   校   卒':
        return 12
    elif x == '高専・短大卒':
        return 14
    elif x == '大学・大学院卒':
        return 16
```

```python
data['education'] = data['学历年龄'].map(edu_trans)
```

```python
def sex_trans(x):
    if x in ['男\u3000 女\u3000 計\n学\u3000 歴   計']:
        return 'delete'
    elif x =='     男\n学\u3000 歴   計':
        return 1
    elif x == '     女\n学\u3000 歴   計':
        return 0
```

```python
data['sex'] = data['学历年龄'].map(sex_trans)
```
